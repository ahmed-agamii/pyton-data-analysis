```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the data from csv file to a pandas Dataframe
raw_mail_data = pd.read_csv('/content/mail_data.csv')
raw_mail_data
```

```python
# replace the null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
mail_data.head()
```

```python
mail_data.shape
```

```python
# replace the null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

```python
mail_data.loc[mail_data['Category']=='spam','Category']=0
mail_data.loc[mail_data['Category']=='ham','Category']=1
```

```python
mail_data.head()
```

```python
#separeting data
X = mail_data['Message']
Y = mail_data['Category']
```

```python
X.head()
```

```python
Y.head()
```

```python
#splitting the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

# feature Extraction

```python
# transform the text data into feature vector can be used in logestic regression
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english')
X_train_feature = feature_extraction.fit_transform(X_train)
X_test_feature = feature_extraction.transform(X_test)

#convert Y into integer

Y_train = Y_train.astype(int)
Y_test = Y_test.astype(int)
```

```python
print(X_train)
```

```python
print(X_train_feature)
```

# training model

```python
model = LogisticRegression()

model.fit(X_train_feature, Y_train)

```

```python
X_train_pred = model.predict(X_train_feature)
acuracy_train_data = accuracy_score(Y_train, X_train_pred)
print('accuracy score of train data =', acuracy_train_data*100,'%')
```

```python
X_test_pred = model.predict(X_test_feature)
acuracy_test_data = accuracy_score(Y_test, X_test_pred)
print('accuracy score of test data =', acuracy_test_data*100,'%')
```

# Building a Predictive System

```python
input_mail=["You've won a prize for free!"]
input_mail_feature = feature_extraction.transform(input_mail)
prediction = model.predict(input_mail_feature)
if prediction[0]== 1 :
  print('ham mail')
else:
  print('spam mail')
```
